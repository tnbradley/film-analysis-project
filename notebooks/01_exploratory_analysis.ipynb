This notebook explores IMDb movie data (2000–2024) to uncover insights about film performance, genre trends, and audience preferences.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print('✅ Libraries imported successfully')
```

## 1. Load the Dataset

```python
# Load the cleaned IMDb movie dataset
df = pd.read_csv('../data_raw/imdb_movies_cleaned.csv')

print(f"Total movies: {len(df):,}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nFirst few rows:")
df.head()
```

## 2. Quick Profile

```python
# Numeric columns summary
print("=== NUMERIC COLUMNS ===")
print(df[['startYear', 'runtimeMinutes', 'averageRating', 'numVotes']].describe())

print("\n=== CATEGORICAL COLUMNS ===")
print(f"\nTop Genres:")
print(df['primaryGenre'].value_counts().head(10))

print(f"\nRating Categories:")
print(df['ratingCategory'].value_counts())

print(f"\nPopularity Tiers:")
print(df['popularityTier'].value_counts())
```

## 3. Genre Performance Analysis

*Research Question: Do certain genres consistently overperform?*

```python
# Average rating by genre (top 10 genres only)
top_genres = df['primaryGenre'].value_counts().head(10).index
genre_stats = df[df['primaryGenre'].isin(top_genres)].groupby('primaryGenre').agg({
    'averageRating': ['mean', 'median', 'count'],
    'numVotes': 'mean'
}).round(2)

genre_stats.columns = ['Avg_Rating', 'Median_Rating', 'Movie_Count', 'Avg_Votes']
genre_stats = genre_stats.sort_values('Avg_Rating', ascending=False)

print("=== GENRE PERFORMANCE ===")
print(genre_stats)
```

```python
# Visualize: Average rating by genre
plt.figure(figsize=(12, 6))
genre_stats_sorted = genre_stats.sort_values('Avg_Rating', ascending=True)

plt.barh(genre_stats_sorted.index, genre_stats_sorted['Avg_Rating'], color='steelblue')
plt.xlabel('Average IMDb Rating')
plt.title('Average Rating by Genre (Top 10 Genres)', fontsize=14)
plt.xlim(5, 8)

# Add value labels
for i, v in enumerate(genre_stats_sorted['Avg_Rating']):
    plt.text(v + 0.05, i, f'{v:.2f}', va='center')

plt.tight_layout()
plt.savefig('../outputs/genre_ratings.png', dpi=150)
plt.show()
```

## 4. Popularity vs. Quality Analysis

*Research Question: Do popular movies always have high ratings?*

```python
# Correlation between popularity (numVotes) and rating
correlation = df['numVotes'].corr(df['averageRating'])
print(f"Correlation between vote count and rating: {correlation:.3f}")

# Scatter plot (sample for performance)
sample_df = df.sample(n=min(5000, len(df)), random_state=42)

plt.figure(figsize=(12, 6))
plt.scatter(sample_df['averageRating'], sample_df['numVotes'], alpha=0.3, s=10)
plt.xlabel('Average Rating')
plt.ylabel('Number of Votes (Popularity)')
plt.title('Popularity vs. Rating: Are Popular Movies Better?')
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.savefig('../outputs/popularity_vs_rating.png', dpi=150)
plt.show()
```

## 5. Hidden Gems: High Rating, Low Awareness

*Strategic Value: Undiscovered films for targeted marketing*

```python
# Define "Hidden Gems": Rating >= 7.5, Votes < 5,000
hidden_gems = df[(df['averageRating'] >= 7.5) & (df['numVotes'] < 5000)]

print(f"Total Hidden Gems: {len(hidden_gems):,}")
print(f"\nTop Hidden Gems by Rating:")

gems_display = hidden_gems.nlargest(10, 'averageRating')[['primaryTitle', 'startYear', 'primaryGenre', 'averageRating', 'numVotes']]
print(gems_display.to_string(index=False))
```

```python
# Hidden gems by genre
print("\nHidden Gems by Genre:")
print(hidden_gems['primaryGenre'].value_counts().head(10))
```

## 6. Decade Trends

*Research Question: How has film quality changed over time?*

```python
# Average rating by decade
decade_stats = df.groupby('decade').agg({
    'averageRating': 'mean',
    'numVotes': 'mean',
    'primaryTitle': 'count'
}).round(2)

decade_stats.columns = ['Avg_Rating', 'Avg_Votes', 'Movie_Count']
print(decade_stats)
```

```python
# Visualize decade trends
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'steelblue'
ax1.set_xlabel('Decade')
ax1.set_ylabel('Average Rating', color=color)
ax1.plot(decade_stats.index, decade_stats['Avg_Rating'], color=color, marker='o', linewidth=2)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'coral'
ax2.set_ylabel('Movie Count', color=color)
ax2.bar(decade_stats.index, decade_stats['Movie_Count'], color=color, alpha=0.3, width=5)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Film Quality vs. Production Volume by Decade', fontsize=14)
plt.savefig('../outputs/decade_trends.png', dpi=150)
plt.show()
```

Export cleaned data for dashboard creation.
